# [ONLY ONE MANDATORY] Challenge 1
#### In this challenge we will use the `Temp_States`  dataset. 

#### First import it into a dataframe called `temp`.

```python
temp_df = pd.read_csv('Temp_States.csv')
```

#### Print `temp`.

```python
temp_df.head()
```

#### Explore the data types of the *temp* dataframe. What types of data do we have? Comment your result.

```python
temp_df.dtypes
```

```python
temp_df.shape
```

```python
"""
The dataset contains information about cities, states (STR), and temperatures (FLOAT) in the US.
It is a small ds with 7 rows and 3 columns.
"""
```

#### Select the rows where state is New York.

```python
ny_state = temp_df[temp_df['State'] == 'New York']
ny_state
```

#### What is the average temperature of cities in New York?

```python
avg_temp_ny = ny_state['Temperature'].mean()
avg_temp_ny
```

#### Which states and cities have a temperature above 15 degrees Celsius?

```python
above_15_temp = temp_df[temp_df['Temperature'] > 15]

display(above_15_temp[['State', 'City']])
```

#### Now, return only the cities that have a temperature above 15 degrees Celsius.

```python
cities_above_15_temp = above_15_temp['City']

display(cities_above_15_temp)
```

#### Which cities have a temperature above 15 degrees Celcius and below 20 degrees Celsius?

**Hint**: First, write the condition. Then, select the rows.

```python
condition = (temp_df['Temperature'] > 15) & (temp_df['Temperature'] < 20)

cities_between_15_and_20_temp = temp_df.loc[condition, 'City']

display(cities_between_15_and_20_temp)
```

#### Find the mean and standard deviation of the temperature of each state.

```python
state_temp_stats = temp_df.groupby('State')['Temperature'].agg(['mean', 'std'])

display(state_temp_stats)
```

# [ONLY ONE MANDATORY]  Challenge 2

#### Load the `employees` dataset into a dataframe. Call the dataframe `employees`.

```python
employees_df = pd.read_csv('Employee.csv')
employees_df.head()
```

#### Explore the data types of the `employees` dataframe. Comment your results.

```python
employees_df.shape
```

```python
employees_df.info
```

```python
employees_df.dtypes
```

```python
"""
The df contains columns with object (string) data type for Name, Department, Education, Gender, and Title, 
and int64 data type for Years and Salary.
The df contains 9 rows and 7 columns
"""
```

#### What's the average salary in this company?

```python
avg_salary = employees_df['Salary'].mean()
avg_salary
```

#### What's the highest salary?

```python
max_salary = employees_df['Salary'].max()
max_salary
```

#### What's the lowest salary?

```python
min_salary = employees_df['Salary'].min()
min_salary
```

#### Who are the employees with the lowest salary?

```python
employees_lowest_salary = employees_df[employees_df['Salary'] == min_salary]
display(employees_lowest_salary[['Name', 'Salary']])
```

#### Find all the information about an employee called David.

```python
employee_david_info = employees_df[employees_df['Name'] == 'David']
employee_david_info
```

#### Could you return only David's salary?

```python
david_salary = employee_david_info['Salary']
david_salary
```

#### Print all the rows where job title is associate.

```python
associate_jobs = employees_df[employees_df['Title'] == 'associate']
associate_jobs
```

#### Print the first 3 rows of your dataframe.
**Tip**: There are 2 ways to do it. Do it both ways.

```python
# Method 1
employees_df.head(3)
```

```python
# Method 2
employees_df[:3]
```

#### Find the employees whose title is associate and whose salary is above 55.

```python
high_paid_associates = employees_df[(employees_df['Title'] == 'associate') & (employees_df['Salary'] > 55)]
high_paid_associates
```

#### Group the employees by number of years of employment. What are the average salaries in each group?

```python
avg_salary_by_years = employees_df.groupby('Years')['Salary'].mean()
avg_salary_by_years
```

####  What is the average salary per title?

```python
avg_salary_per_title = employees_df.groupby('Title')['Salary'].mean()
avg_salary_per_title
```

####  Find the salary quartiles.


```python
salary_quartiles = employees_df['Salary'].quantile([0.25, 0.5, 0.75])
salary_quartiles
```

#### Is the mean salary different per gender?

```python
mean_salary_per_gender = employees_df.groupby('Gender')['Salary'].mean()
mean_salary_per_gender
```

#### Find the minimum, mean and maximum of all numeric columns for each company department.



```python
department_stats = employees_df.groupby('Department').agg({'Years': ['min', 'mean', 'max'],
                                                           'Salary': ['min', 'mean', 'max']})

department_stats.columns = ['Min_Years', 'Mean_Years', 'Max_Years', 'Min_Salary', 'Mean_Salary', 'Max_Salary']

department_stats
```

#### Bonus Question:  for each department, compute the difference between the maximum and the minimum salary.
**Hint**: try using `agg` or `apply` combined with `lambda` functions.

```python
salary_diff = lambda x: x.max() - x.min()

dpt_salary_diff = employees_df.groupby('Department')['Salary'].agg(salary_diff)

dpt_salary_diff.rename('Salary_Difference', inplace=True)

dpt_salary_diff
```

####  What is the average purchase price?

#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
customers_spain = orders[orders['Country'] == 'Spain']['CustomerID'].unique()
customers_spain
```

#### How many customers do we have in Spain?

```python
num_customers_spain = len(orders[orders['Country'] == 'Spain']['CustomerID'].unique())
num_customers_spain
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
customer_purchase_qties = orders.groupby('CustomerID')['Quantity'].sum()

customers_50_items = customer_purchase_qties[customer_purchase_qties > 50].index

customers_50_items
```

#### Select orders from Spain that include more than 50 items.

```python
orders_spain = orders[orders['Country'] == 'Spain']

orders_50_items_spain = orders_spain[orders_spain['Quantity'] > 50]

orders_50_items_spain
```

#### Select all free orders.

```python
free_orders = orders[orders['amount_spent'] == 0]

free_orders
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
lunch_bag_orders = orders[orders['Description'].str.startswith('lunch bag')]

lunch_bag_orders
```

#### Select all `lunch bag` orders made in 2011.

```python
lunch_bag_orders_2011 = orders[(orders['Description'].str.startswith('lunch bag')) & (orders['year'] == 2011)]

lunch_bag_orders_2011
```

#### Show the frequency distribution of the amount spent in Spain.

```python
import matplotlib.pyplot as plt

orders_spain = orders[orders['Country'] == 'Spain']

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(orders_spain['amount_spent'], bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Amount Spent in Spain')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Select all orders made in the month of August.

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
orders_august = orders[orders['month'] == 8]

orders_by_country_august = orders_august['Country'].value_counts()

orders_by_country_august
```

#### What's the  average amount of money spent by country?

#### What's the most expensive item?

#### What is the average amount spent per year?
